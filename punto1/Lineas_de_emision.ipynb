```python
#importo los datos con el conteo de electrones como función de energía

data = loadtxt("energy_counts.dat")
```

```python
#separo los datos de las columnas en variables y tomo el numero total de datos

energy = data[:,0] #x
n_counts = data[:,1] #y
numData = shape(data)[0]
```

```python
#grafico los datos que entran
scatter(energy, n_counts)
title("$Conteo$ $de$ $electrones$ $como$ $funcion$ $de$ $la$ $energia$", size = 30)
xlabel("$E$", size = 10)
ylabel("$N(E)$",rotation='horizontal', size = 10)
```

```python
#modelo que voy a sacar

def modelo(E,A,B,E0,s,alpha):
   return A*(E**alpha) + B*exp(-((E-E0)/(sqrt(2)*s))**2)
```

```python
#chi2 para encontrar el mas probable

def likelihood (y_obs, y_model):
    chi_squared = sum((n_counts - y_model)**2)
    return chi_squared
```

```python
alpha_walk = empty((0))
A_walk = empty((0))
B_walk = empty((0))
E0_walk = empty((0))
s_walk = empty((0))
chi_2 = empty((0))


#Inicializar las variables (los valores de inicializacion fueron encontrados a partir de la solucion analitica)
A_walk = append(A_walk,10**16)
B_walk =append(B_walk,2000)
E0_walk = append(E0_walk,1400)
s_walk = append(s_walk,100)
alpha_walk = append(alpha_walk,-4)

y_ini = modelo(energy,A_walk[0],B_walk[0],E0_walk[0],s_walk[0],alpha_walk[0])
chi_2 = append(chi_2,likelihood(n_counts, y_ini))
```

```python
n_iterations = 2000 #Numero de iteraciones al metodo a realizar

for i in range(n_iterations):
    
    alpha_prime = random.normal(alpha_walk[i], 0.1) 
    A_prime = random.normal(A_walk[i], 0.1)
    s_prime = random.normal(s_walk[i], 0.1) 
    B_prime = random.normal(B_walk[i], 0.1)
    E0_prime = random.normal(E0_walk[i], 0.1) 
    
    y_init = modelo(energy, A_walk[i],B_walk[i],E0_walk[i],s_walk[i],alpha_walk[i])
    y_prime = modelo(energy, A_prime,B_prime,E0_prime,s_prime,alpha_prime)
    
    chi_prime = likelihood(n_counts,y_prime)
    chi_init = likelihood(n_counts,y_init)
    
    psi = chi_prime/chi_init
    if(psi>=1.0):
        alpha_walk  = append(alpha_walk,alpha_prime)
        A_walk  = append(A_walk,A_prime)
        E0_walk  = append(E0_walk,E0_prime)
        s_walk  = append(s_walk,s_prime)
        B_walk  = append(B_walk,B_prime)
        chi_2 = append(chi_2,likelihood(n_counts,y_prime))
    else:
        beta = random.random()
        if(beta<=psi):
            alpha_walk  = append(alpha_walk,alpha_prime)
            A_walk  = append(A_walk,A_prime)
            E0_walk  = append(E0_walk,E0_prime)
            s_walk  = append(s_walk,s_prime)
            B_walk  = append(B_walk,B_prime)
            chi_2 = append(chi_2,likelihood(n_counts,y_prime))
        else:
            alpha_walk  = append(alpha_walk,alpha_walk[i])
            A_walk  = append(A_walk,A_walk[i])
            E0_walk  = append(E0_walk,E0_walk[i])
            s_walk  = append(s_walk,s_walk[i])
            B_walk  = append(B_walk,B_walk[i])
            chi_2 = append(chi_2,chi_init)

print alpha_prime, A_prime, B_prime, E0_prime, s_prime
```

```python
#Encuentro los valores promedio

ave_A=average(A_walk)
ave_B=average(B_walk)
ave_E0=average(E0_walk)
ave_s=average(s_walk)
ave_alpha=average(alpha_walk)

print ave_A,ave_B,ave_E0,ave_s,ave_alpha
```

```python
#Encuentro los valores de las constantes

min_likelihood_id = argmin(chi_2)
best_A = A_walk[min_likelihood_id]
best_B = B_walk[min_likelihood_id]
best_E0 = E0_walk[min_likelihood_id]
best_s = s_walk[min_likelihood_id]
best_alpha = alpha_walk[min_likelihood_id]
```

```python
#Comparo los datos de entrada con los valores que me dan para revisar si son correctos

bestfit = modelo(energy, best_A, best_B, best_E0, best_s, best_alpha)

plot(energy,bestfit,'.', label = 'valores de constantes')
plot(energy,n_counts, 'r', label = 'datos iniciales')

title("$Conteo$ $de$ $electrones$ $como$ $funcion$ $de$ $la$ $energia$", size = 20)
xlabel("$E$", size = 10)
ylabel("$N(E)$",rotation='horizontal', size = 10)
legend()
```

```python
#saco el error del valor con respecto al valor promedio

err_A=abs(best_A-ave_A)
err_B=abs(best_B-ave_B)
err_E0=abs(best_E0-ave_E0)
err_s=abs(best_s-ave_s)
err_alpha=abs(best_alpha-ave_alpha)

print err_A, err_B, err_E0, err_s, err_alpha
```
